```python
from datetime import datetime
from dateutil.relativedelta import relativedelta


def calculate_remaining_lease(lease_year):

    CURRENT_DATE = datetime.today()


    # Create lease start date
    lease_start_date = datetime(
        int(lease_year),
        1,
        1
    )


    # HDB lease is 99 years
    lease_end_date = (
        lease_start_date
        + relativedelta(years=99)
    )


    # Calculate difference
    remaining = relativedelta(
        lease_end_date,
        CURRENT_DATE
    )


    # Handle expired lease
    if remaining.years < 0:
        return "0 years 0 months"


    return (
        f"{remaining.years} years "
        f"{remaining.months} months"
    )
```

```python
import re

def create_resale_identifier(row):

    block_digits = re.sub(r"\D", "", str(row["block"]))

    block_digits = block_digits[:3].zfill(3)

    avg_digits = str(int(row["avg_price"]))[:2]

    month = row["month"][-2:]

    town_initial = row["town"][0]

    return f"S{block_digits}{avg_digits}{month}{town_initial}"
```

```python
clean_df["Resale Identifier"] = clean_df.apply(
    create_resale_identifier,
    axis=1
)
```

```python
import hashlib

def hash_identifier(df):

    df = df.copy()

    df["Hashed Identifier"] = (
        df["Resale Identifier"]
        .apply(
            lambda x: hashlib.sha256(
                x.encode("utf-8")
            ).hexdigest()
        )
    )

    return df
```
